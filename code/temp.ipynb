```python
import numpy as np
import matplotlib.pyplot as plt
from math import comb
```

```python
def num_ass(d):
    total_sum = 0
    for k in range(1, d):
        dk = comb(d, k)
        incomb = 0
        for j in range(1, d-k+1):
            incomb += comb(d-k, j)
        total_sum += dk * incomb
    return total_sum
```

```python
x = list(range(2,15))
y = [num_ass(i) for i in x]
```

```python
plt.xticks(ticks=x)
plt.plot(x,y,'k--')
plt.plot(x,y,'ro')
plt.title('Number of Rules for an Itemset')
plt.ylabel('Rules')
plt.xlabel('Length of Itemset')
plt.gcf().set_size_inches(10,6)
```

```python
for i in range(2, 10):
    print(f'{i}: {num_ass(i):,}')
```

```python
y = list(range(2, 200))
approx = [x ** (1.5) for x in y]
mst = [x ** 2 for x in y]

fig, ax = plt.subplots()
ax.plot(mst, y, 'r--', label='Approximated MST O(n^1.5)')
ax.plot(approx, y, 'g--', label='Exact MST O(n^2)')
plt.legend()
plt.ylabel('N', rotation=0)
plt.xlabel('Number of Loops')
plt.gcf().set_size_inches(10,6)
plt.show()
```
